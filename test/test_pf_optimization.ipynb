```julia
include("test/test_power_flow.jl")
using NLopt

count = 0 # keep track of # function evaluations

function pf_obj(x::Vector, grad::Vector, mismatch::Function)
    if length(grad) > 0
        g, Jac = mismatch(x, NeedMismatch=true, NeedJac=true)
        grad[:] = g'*Jac
    else
        g = mismatch(x, NeedMismatch=true, NeedJac=false)[1]
    end
    
    global count
    count::Int += 1
    println("f_$count(x)")
    
    return (0.5*g'*g)[1]
end

opt = Opt(:LN_NEWUOA , nx)
#lower_bounds!(opt, [-Inf, 0.])
xtol_rel!(opt,1e-4)

min_objective!(opt, (x,y) -> pf_obj(x,y,mismatch))

(minf,minx,ret) = optimize(opt, x0)
println("got $minf at $minx after $count iterations (returned $ret)")
```

```julia
using NLopt

count = 0 # keep track of # function evaluations

function myfunc(x::Vector, grad::Vector, alaki)
    if length(grad) > 0
        grad[1] = 0
        grad[2] = 0.5/sqrt(x[2])
    end

    global count
    count::Int += 1
    println("f_$count($x)")

    sqrt(x[2])
end

function myconstraint(x::Vector, grad::Vector, a, b)
    if length(grad) > 0
        grad[1] = 3a * (a*x[1] + b)^2
        grad[2] = -1
    end
    (a*x[1] + b)^3 - x[2]
end

opt = Opt(:LD_MMA, 2)
lower_bounds!(opt, [-Inf, 0.])
xtol_rel!(opt,1e-4)

min_objective!(opt, (x,y) -> myfunc(x,y,3))
inequality_constraint!(opt, (x,g) -> myconstraint(x,g,2,0), 1e-8)
inequality_constraint!(opt, (x,g) -> myconstraint(x,g,-1,1), 1e-8)

(minf,minx,ret) = optimize(opt, [1.234, 5.678])
println("got $minf at $minx after $count iterations (returned $ret)")
```
